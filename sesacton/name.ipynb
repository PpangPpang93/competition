```python
import numpy as np
import torch
from torchvision import models, transforms
import torch.nn as nn

from PIL import Image
import matplotlib.pyplot as plt
```

```python
class_names = ['Accessories', 'Backpack', 'Bag', 'Bag_hand', 'Bag_hip', 'Bag_shoulder', 'Clothing_pants', 'Clothing_top', 'Earphones', 'Hat', 'Labtop', 'Labtop_LG', 'Labtop_Macbook', 'Labtop_Samsung', 'Pattern_check', 'Pattern_flower', 'Pattern_stripe', 'Phone_galaxy', 'Phone_iPhone', 'Shoes', 'Sneaker', 'Wallet', 'Watch']
CKPPATH = 'D:/sesac/model/ckp/obj_3.pth'
```

```python
model = models.resnet50(pretrained=True)

class_num = 37
class_num_ftrs = model.fc.in_features
model.fc = nn.Linear(class_num_ftrs, class_num)

device = torch.device('cuda:0')
model.to(device)

img_path = 'C:/Users/ppang/Desktop/sample/지갑.jpg'

ckp = torch.load(CKPPATH)
model.load_state_dict(ckp['model_state_dict'])
```

```python
transforms_test = transforms.Compose([
    transforms.Resize((224, 224)),
    transforms.ToTensor(),
    transforms.Normalize([0.485, 0.456, 0.406], [0.229, 0.224, 0.225])
])
image = Image.open(img_path)
image = transforms_test(image).unsqueeze(0).to(device)
```

```python
with torch.no_grad():
    outputs = model(image)
    _, preds = torch.topk(outputs, 5)
    
    print(class_names[preds[0][0]], class_names[preds[0][1]], class_names[preds[0][2]], class_names[preds[0][3]], class_names[preds[0][4]])
```
