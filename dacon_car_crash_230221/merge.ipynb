```python
import os
import pandas as pd
```

```python
path = '/data/home/ubuntu/workspace/dacon/data/'
```

```python
weather_snow = pd.read_csv(path + 'test_weather_snow_230310.csv')
weather_rain = pd.read_csv(path + 'test_weather_rain_230310.csv')
timing = pd.read_csv(path + 'test_timing_230309.csv')
crash = pd.read_csv(path + 'test_crash_ego_230310.csv')
```

```python
total_df = crash.merge(weather_snow).merge(weather_rain).merge(timing).drop(columns=['aug','video_path'])
total_df
```

```python
total_df['weather'] = ''
total_df
```

```python
for i in range(len(total_df)):
    if (total_df.loc[i, 'crash_ego']==0):
        total_df.loc[i, 'weather'] = 0
        
    if (total_df.loc[i, 'snow']==1):
        total_df.loc[i, 'weather'] = 1
        
    elif (total_df.loc[i, 'rain']==1):
        total_df.loc[i, 'weather'] = 2
        
    else:
        total_df.loc[i, 'weather'] = 0
```

```python
total_df = total_df[['sample_id','crash_ego','weather','timing']]
total_df
```

```python
# total_df.to_excel('./data/temp_0309.xlsx', index=False)
```

- crash_ego : 0(무사고), 1(사고 + 내가 사고), 2(사고+ 남이 사고)
- weather : 0(정상), 1(눈), 2(비)
- timing : 0(낮), 1(밤)

```python
total_df['pred'] = ''
for i in range(len(total_df)):
    if (total_df.loc[i, 'crash_ego'] == 0):
        total_df.loc[i, 'pred'] = 0
        
    elif ((total_df.loc[i, 'crash_ego'] == 1) & (total_df.loc[i, 'weather'] == 0) & (total_df.loc[i, 'timing'] == 0)):
        total_df.loc[i, 'pred'] = 1
        
    elif ((total_df.loc[i, 'crash_ego'] == 1) & (total_df.loc[i, 'weather'] == 0) & (total_df.loc[i, 'timing'] == 1)):
        total_df.loc[i, 'pred'] = 2
        
    elif ((total_df.loc[i, 'crash_ego'] == 1) & (total_df.loc[i, 'weather'] == 1) & (total_df.loc[i, 'timing'] == 0)):
        total_df.loc[i, 'pred'] = 3
        
    elif ((total_df.loc[i, 'crash_ego'] == 1) & (total_df.loc[i, 'weather'] == 1) & (total_df.loc[i, 'timing'] == 1)):
        total_df.loc[i, 'pred'] = 4
        
    elif ((total_df.loc[i, 'crash_ego'] == 1) & (total_df.loc[i, 'weather'] == 2) & (total_df.loc[i, 'timing'] == 0)):
        total_df.loc[i, 'pred'] = 5
        
    elif ((total_df.loc[i, 'crash_ego'] == 1) & (total_df.loc[i, 'weather'] == 2) & (total_df.loc[i, 'timing'] == 1)):
        total_df.loc[i, 'pred'] = 6
        
    elif ((total_df.loc[i, 'crash_ego'] == 2) & (total_df.loc[i, 'weather'] == 0) & (total_df.loc[i, 'timing'] == 0)):
        total_df.loc[i, 'pred'] = 7
        
    elif ((total_df.loc[i, 'crash_ego'] == 2) & (total_df.loc[i, 'weather'] == 0) & (total_df.loc[i, 'timing'] == 1)):
        total_df.loc[i, 'pred'] = 8
        
    elif ((total_df.loc[i, 'crash_ego'] == 2) & (total_df.loc[i, 'weather'] == 1) & (total_df.loc[i, 'timing'] == 0)):
        total_df.loc[i, 'pred'] = 9
        
    elif ((total_df.loc[i, 'crash_ego'] == 2) & (total_df.loc[i, 'weather'] == 1) & (total_df.loc[i, 'timing'] == 1)):
        total_df.loc[i, 'pred'] = 10
        
    elif ((total_df.loc[i, 'crash_ego'] == 2) & (total_df.loc[i, 'weather'] == 2) & (total_df.loc[i, 'timing'] == 0)):
        total_df.loc[i, 'pred'] = 11
        
    elif ((total_df.loc[i, 'crash_ego'] == 2) & (total_df.loc[i, 'weather'] == 2) & (total_df.loc[i, 'timing'] == 1)):
        total_df.loc[i, 'pred'] = 12
    
```

```python
submission = pd.read_csv(path + 'sample_submission.csv')
submission['label'] = total_df['pred']
submission.to_csv(path + '230310_submission2.csv', index=False)
submission
```
